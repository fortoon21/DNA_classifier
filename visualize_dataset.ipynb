```python
import torch
from sklearn.datasets import make_classification
import numpy as np
```

```python
pre_X, y = make_classification(n_samples=2000, n_classes=2, n_informative=3, n_redundant=0, 
                    n_features=3, random_state=123)
X=pre_X + abs(np.min(pre_X)) + np.array([1,1,1])
X=np.around(X*0.1, 4)
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import animation
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(10,10))
ax = Axes3D(fig)
colors = 'b', 'r'
labels = 'Group1', 'Group2'

for i, c, label in zip(range(len(labels)), colors, labels):
    ax.scatter(X[y==i,0], X[y==i,1], X[y==i,2], s=50, c=c, label=label, alpha=0.5)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

fig.legend()
```

```python
def rotate(angle):
     ax.view_init(azim=angle)

angle = 3
ani = animation.FuncAnimation(fig, rotate, frames=np.arange(0, 360, angle), interval=50)
ani.save('rotate_3D.gif', writer=animation.PillowWriter(fps=20))
```
