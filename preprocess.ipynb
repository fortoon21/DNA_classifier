```python
rna_exp = pd.read_csv('RNASeqV2.geneExp.T.txt', sep='\t')
```

```python
labels=pd.read_csv('merged_sample_quality_annotations.tsv', sep='\t')
```

```python
RNA_labels=labels[labels['platform']=='IlluminaHiSeq_RNASeqV2']
```

```python
index=[]
for canc in pd.DataFrame(RNA_labels['cancer type']).value_counts().index:
    index.append(canc[0])
index
```

```python
dictionary = {}
for i, v in zip(range(len(index)), index):
    dictionary[v]=i
print(dictionary)
```

```python
data={}

for i, barcode in enumerate(RNA_labels['aliquot_barcode']):
    if barcode in rna_exp['gene_id'].values :
        data[barcode]=RNA_labels.iloc[i]['cancer type']
    else :
        data[barcode]='NaN'
```

```python
df=pd.DataFrame([data]).T
df2=df.reset_index(level=0)
df2.columns=['gene_id', 'cancer type']
df2.head()
```

```python
df3=pd.merge(rna_exp, df2, on='gene_id')
```

```python
df3.to_csv('RNASeqV2.geneExp.df.tsv', index=False, sep="\t")
```

```python
df3['cancer type'].value_counts()
```

```python
SKCM=df3[df3['cancer type']=='SKCM']
PRAD=df3[df3['cancer type']=='PRAD']
LUAD=df3[df3['cancer type']=='LUAD']
```

```python
tiny_data=pd.DataFrame(pd.concat([SKCM, PRAD, LUAD]), columns=columns)
tiny_data
```

```python
tiny_data.to_csv('tiny_data.df.tsv', index=False, sep="\t")
```
